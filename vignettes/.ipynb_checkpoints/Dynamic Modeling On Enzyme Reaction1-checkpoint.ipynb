## Basic Enzyme Reaction

The basic enzyme reaction is a well-known example of a (bio)chemical reaction network. It describes a system where a substrate S reacts with an enzyme E to form a complex C which is then converted into a product P and the enzyme.

Differential equations for the concentrations of each molecule in the system can be obtained by application of the law of mass-action, which states that the rate of a reaction is proportional to the product of the concentration of the reactants.

## Constraints

There are two constraints, first one is based on conservation laws, that C(t) + E(t) should be constant.

Another constraint is that, when we calculate dS(t) - dE(t), we can get $k_0 - k_2C(t)$, therefore, to make the system converge, $C(t) = \frac{k_0}{k_2}$

### We will make SCM by pyro and use the ODE above as data generating procedure.

```python
import random
import torch
import pyro
import pyro.distributions as dist
from pyro.infer import EmpiricalMarginal
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
def model_ode():
    
    #initialize rates & states
    rates = [0.04, 0.03, 0.1, 0.11, 0.05]
    inner_states = [8,1,1,1]
    S, E, C, P = inner_states

    k_0, k_minus1, k_1, k_2, k_3 = rates
    k_0 = torch.tensor(k_0)
    k_minus1 = torch.tensor(k_minus1)
    k_1 = torch.tensor(k_1)
    k_2 = torch.tensor(k_2)
    k_3 = torch.tensor(k_3)
            
    state_list = []
    
    
    #implement ODEs
    for _ in range(500):
        dE = (-k_1) * S * E + (k_minus1 + k_2) * C
        dS = k_0 - (k_1 * S * E) + (k_minus1 * C)
        dC = k_1 * S * E - (k_minus1 + k_2) * C
        dP = k_2 * C - k_3 * P
    
#         N_S = pyro.sample('pN_S', dist.Normal(0., 0.001))
#         N_E = pyro.sample('pN_E', dist.Normal(0., 0.001))
#         N_C = pyro.sample('pN_C', dist.Normal(0., 0.001))
#         N_P = pyro.sample('pN_P', dist.Normal(0., 0.001))
        
        N_S = N_E = N_C = N_P = 0.
        S = pyro.sample('pS', dist.Delta(S + dS + N_S))
        E = pyro.sample('pE', dist.Delta(E + dE + N_E))
        C = pyro.sample('pC', dist.Delta(C + dC + N_C))
        P = pyro.sample('pP', dist.Delta(P + dP + N_P))
        
        inner_states = [S, E, C, P]
        state_list.append(inner_states)
        
    return state_list
```

### Simulation on different intervention sets

```python
def make_title(do_value_dict):
    string=''
    for i,x in zip(do_value_dict.keys(),do_value_dict.values()):
        string+='do({}={})'.format(i[1:],'%.2f' %x)
        string+=' '
    return string
```

```python
# do model
def docheck(do_value_dict):
    model_doC = pyro.do(model_ode, data=do_value_dict)
    sample_doC = model_doC()
    state_array = np.asarray(sample_doC)
    flag=1
    
    #remove outliers,
    for i in range(len(state_array)):
        tmp=state_array[i]
        for j in tmp:
            if j<0 or j>100:
                flag=0
                break
        if flag==0:
            break
        
    state_array=state_array[:i]
    plt.figure()
    plt.plot(state_array[:,0])
    plt.plot(state_array[:,1])
    plt.plot(state_array[:,2])
    plt.plot(state_array[:,3])
    plt.title(make_title(do_value_dict))
    plt.legend(['S','E','C','P'])
    plt.show()
```

### Simulation without intervention

```python
state_list = model_ode()
state_array = np.asarray(state_list)
plt.plot(state_array[:,0])
plt.plot(state_array[:,1])
plt.plot(state_array[:,2])
plt.plot(state_array[:,3])
plt.legend(['S','E','C','P'])
plt.show()
```

## Intervention on 1 variable

### Simulation with intervention on S, E, P

```python
docheck({'pS': torch.tensor(2)})
docheck({'pE': torch.tensor(2)})
docheck({'pP': torch.tensor(2)})
```

### Simulation with intervention on C $(C\neq \frac{k_0}{k_2})$, the system does not converge

```python
docheck({'pC': torch.tensor(2)})
```

### Simulation with intervention on C $(C = \frac{k_0}{k_2})$, the system does converge

```python
k_0 = 0.04
k_2 = 0.11
docheck({'pC': torch.tensor(k_0/k_2)})
```

## Intervention on 2 variables

### Simulation with intervention on 2 variables, the systems converge

```python
docheck({'pC': torch.tensor(2), 'pS': torch.tensor(1)})
docheck({'pC': torch.tensor(2), 'pE': torch.tensor(1)})
docheck({'pS': torch.tensor(2), 'pP': torch.tensor(1)})
docheck({'pS': torch.tensor(2), 'pE': torch.tensor(1)})
docheck({'pE': torch.tensor(2), 'pP': torch.tensor(1)})
```

### Simulation with intervention on C and P $(C\neq \frac{k_0}{k_2})$, the system does not converge

```python
docheck({'pC': torch.tensor(2), 'pP': torch.tensor(1)})
```

### Simulation with intervention on C and P $(C = \frac{k_0}{k_2})$, the system does converge

```python
k_0 = 0.04
k_2 = 0.11
docheck({'pC': torch.tensor(k_0/k_2), 'pP': torch.tensor(1)})
```

To explain why we do intervention on C and C,P will make the system not converge, we go back to the second constraint.

### Simulation with intervention on 3 variables, the systems always converge

```python
docheck({'pS': torch.tensor(3), 'pE': torch.tensor(2), 'pC': torch.tensor(1)})
docheck({'pS': torch.tensor(3), 'pE': torch.tensor(2), 'pP': torch.tensor(1)})
docheck({'pS': torch.tensor(3), 'pC': torch.tensor(2), 'pP': torch.tensor(1)})
docheck({'pE': torch.tensor(3), 'pC': torch.tensor(2), 'pP': torch.tensor(1)})
```

### Simulation with intervention on all variables, the system certainly converges

```python
docheck({'pS': torch.tensor(4), 'pE': torch.tensor(2), 'pC': torch.tensor(3), 'pP': torch.tensor(1)})
```

### Our simulation proves the contraint model describe in Blom & Mooij's paper. The system has constraint set {C}, {C,P}. In these two cases, if the intervention on C is not equivalent to $\frac{k_0}{k_2}$, the equilibrium will not achieve.
